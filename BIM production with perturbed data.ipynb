The company Best international Machines(BIM) realizes that a 1% fraction of the copper always gets wasted while producing both types of microchips, more specifically 1% of the required amount. This means that it actually takes 4.04gr of copper to produce a logic chip and 
 2.02gr of copper to produce a memory chip. Solving this require much safer approach is to explicitly define the two decision variables to be non-negative integers, thus transforming the original into the following mixed-integer linear optimization (MILO) problem

```python
from amplpy import AMPL,ampl_notebook 
ampl=ampl_notebook(modules=['highs','gurobi'],
                  license_uuid='fb5195d0-ea81-4d48-86fb-5c8008b32fd0')
```

```python
%%writefile BIM_perturbed_MILO.mod
var x1 integer>=0;
var x2 integer>=0;

maximize profit:
    12*x1+9*x2;
s.t. silicon: x1<=1000;
s.t. germanium: x2<=1500;
s.t. plastic: x1+x2<=1750; 
s.t. copper: 4.04*x1+2.02*x2<=4800;
```

```python
m=AMPL()
m.read("BIM_perturbed_MILO.mod")
m.solve(solver=SOLVER)
assert m.solve_result=='solved',m.solve_result

print(
    "x = ({:.3f}, {:.3f}), optimal value = {:.3f}".format(
        m.var["x1"].value(), m.var["x2"].value(), m.obj["profit"].value()
    )
)
```
